```python
  
# 이미지를 Computer Vision API로 분석하여 특징 벡터 추출  
with open(image_path, "rb") as image_stream:  
    image_analysis = vision_client.analyze_image_in_stream(  
        image_stream,  
        visual_features=[VisualFeatureTypes.tags]  
    )  
  
# 이미지 분석 결과에서 태그 추출  
image_tags = [tag.name for tag in image_analysis.tags]  
  
# Azure Search에서 이미지 태그를 기반으로 검색  
search_results = search_client.search(search_text=" ".join(image_tags))  
  
# 검색 결과 출력  
for result in search_results:  
    print(f"일치율: {result['@search.score']}")  
    print(f"Index Title: {result['title']}")  
```
